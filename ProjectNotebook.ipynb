# Project Description

This project is aiming to introduce some roles in a recently popular game, which called Animal Crossing. 
In this game,the player character is a human who lived in a village inhabited by various anthropomorphic animals, carrying out various activities such as fishing, bug catching, and fossil hunting. 

New beginners might get confused about the animals' role, species and personality. 
In order to solve that concern, I am developing this Animal Chatbot to introduce some animal's species, role and personality.

I got inspired by one of the example project on the course webpage. Here is my reference: Byungkwon (S19): https://github.com/chrishan0106/cogs18_final. I make some modifications of this project as below. I also carry some functions from A3.

```python

def have_a_chat_AnimalBot():
    """Main function to run Animal bot. """
    # set up the chat environment 
    dics = input('INPUT :\t')

    dics_Str = list_to_string(dics,' ')
    dics = prepare_text(dics_Str)
    start()
    next_msg='Do you want additional information? Type Yes or No'

    chat = True

    #Theses three parameters to controll the following loops 
    counter = 0
    prompt = False
    cross = 0
    person2 = False
    species2 = False 
    role2 = False 
    role = False

    #checking yes or no
    yes_or_no = False
    
    while chat:

        # Get a message from the user
        msg = input('')
        out_msg = None
        msg = prepare_text(msg)

        # Check if the input is a 'role'
        role = is_role(msg)
        # Check if the input is a 'species'
        species = is_species(msg)

        # Check if the input is a 'more'
        is_more = more(msg)

        # Check if the input is a 'list'
        menu = is_List(msg)

        # Check if the input is a 'personality'
        personality = is_personality(msg)

        # Check if the input is one of the personalities
        role_personality = is_personality_animal(msg)


        # Check if the input is one of the names
        role_name = is_role_animal(msg)


        # Check if the input is one of the species 
        role_species = is_species_animal(msg)
        



        # Check for an end msg 
        ##Ends the bot if the user types exit or quit
        outs_msg = []

        
        if end_chat(msg):
            out_msg = random.choice(goodbye)
            chat = False 
            prompt = False
            print(out_msg)
            break


        ##Only runs this elif loop when the input is one of the roles and after the question. 
        elif is_in_list(msg,dics) and counter is 1:

            ##Response with specific personality that the role belongs to 
            if person2 is True:
                out_msg = 'The personality of '+list_to_string(msg,' ')+' is/are '+ get_personality(msg)
                prompt = True 
                person2 = False 
                flag = 1
                print(out_msg)

            ##Response with specific species that the role belongs to 
            elif species2 is True:
                out_msg = 'The species of '+list_to_string(msg,' ')+' is/are '+ get_Species(msg)
                prompt = True 
                species2 = False 
                flag = 1
                print(out_msg)

        ##Shows the list of species for users. 
        elif menu is True and counter is 1:
            out_msg = 'These are the List of Species: '+species_of_animal()+ "."
            prompt = False
            flag = 0
            print(out_msg)

         ##Gives response with the list of roles that belong to the specific personality 
        elif role_personality is True and counter is 1:
            out_msg = 'The animals with personality '+list_to_string(msg,' ')+' is/are ' +get_personality_role(msg)
            flag = 1
            prompt = True
            print(out_msg)


        ##Gives response with the list of characters that belong to the specific species 
        elif role_species is True and counter is 1 and role2 is False: 
            out_msg = 'The role of '+list_to_string(msg,' ')+' is/are '+ get_role_species(msg)
            prompt = True
            flag = 1
            print(out_msg)

        elif role_name is True and counter is 1 and role2 is False:
            out_msg = 'The role with '+list_to_string(msg,' ')+' is/are '+ get_Species(msg)
            prompt = True
            flag = 1
            print(out_msg)
            

        elif role_name is True and counter is 1 and role2 is True:
            out_msg = 'The role with '+list_to_string(msg,' ')+' is/are '+ get_personality(msg)
            prompt = True 
            flag = 1
            print(out_msg)
            

        ##Asks which character's role info user wants to know 
        elif role is True and counter is 0:
            out_msg = "Which Role's personality information are you looking for?"
            role2 = True 
            counter = 1
            prompt = False
            flag = 0
            print(out_msg)

        ##Asks user to type the species they want to find role of. 
        elif species is True and counter is 0 :
            out_msg = 'Type the species or specific role. If you want to see the list of species, type list'
            role2 = False
            prompt = False
            counter = 1
            flag = 0
            print(out_msg)

        ##Asks user to type the personality they want to find characters of. 
        elif personality is True and counter is 0:
            out_msg = "Type the personality. There are lazy, cranky, normal, smug, peppy, jock, snooty."
            role2 = False
            prompt = False
            counter = 1 
            flag = 0
            print(out_msg)


        ##Asks how user wants to find the list of roles 
        elif role is True and counter is 0:
            out_msg = "How do you want to find roles? personality or species?"
            role2 = True 
            prompt = False
            flag = 0
            print(out_msg)
            
        ##Prints error statement if input does not match any of the cases so far
        else:
            if (yes_or_no == False ):
                out_msg = random.choice(unknown_out)
                propmt = False
                print(out_msg)
            elif((list_to_string(msg,'').lower() != "yes") & (list_to_string(msg,'').lower() != "no")):
                out_msg = random.choice(unknown_out)
                print(out_msg)
                continue


        if prompt is True:
            print(next_msg)
            yes_or_no = True
            prompt = False
            continue

        yes_or_no = False


        ##If user wants more information, ask what kind of information they want 
        if is_more is True and counter is 1 and flag is 1:
                out_msg = 'What do you want to know more? Role or species or personality?' 
                prompt = False
                counter = 0
                flag = 0
                print(out_msg)

        ##Ends bot if user doesn't want more information         
        elif  counter is 1 and flag is 1:
            out_msg = random.choice(goodbye)
            chat = False 
            prompt = False
            print(out_msg)
```
